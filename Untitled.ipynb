```python
import numpy as np
import matplotlib.pyplot as plt
x_true = np.linspace(-1,2,100)
y_true = np.tan(np.pi * x_true / 3) + (x_true - 0.5) ** 2

x_2 = np.linspace(-1,1,50)
y_2 = np.tan(np.pi * x_2 / 3) - (x_2 - 0.5) ** 2

plt.plot(x_2, y_2, linewidth=2, color='black')
plt.plot(x_true, y_true, linewidth=2, color='g')

plt.show()
```
